### Movie Recommendation System

## Problem Definition

> Recommeding the movies based on Similarity

## Data

> The Data has been downloaded from Kaggle - https://www.kaggle.com/datasets/ahsanaseer/top-rated-tmdb-movies-10k

## Features

> The columns given in the data are:
* ID: Movie ID number on the website
* title: Movie name
* genre: Movie genre (crime, adventure, etc.)
* original_language: Original language in which the movie is released
* overview: Summary of the movie
* popularity: Movie Popularity
* release_date: Movie release date
* vote_average: Movie vote average
* vote_count: Movie vote count

```python
movies = movies[["id", "title", "overview", "genre"]]
```

```python
movies
```

```python
movies.genre
```

```python
movies["tags"] = movies["overview"] + movies["genre"]
```

```python
movies
```

```python
new_data = movies.drop(columns = ["overview", "genre"])
```

```python
new_data
```

```python
# Count vectorizer - convert to machine language
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words="english")
```

```python
cv
```

```python
vector = cv.fit_transform(new_data["tags"].values.astype("U")).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
# Angle is high - less similar, angle is low - more similar
similarity
```

```python
new_data[new_data["title"] == "The Godfather"].index[0]
```

```python
new_data[new_data["title"] == "Domino"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse = True, key = lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    index = new_data[new_data["title"] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```

```python
recommend("Captain America")
```

```python
recommend("Batman Begins")
```
